# Exercise 3: Creating new feature

1. Create a new feature 'AgeGroup' based on the 'age' column with the following labels: 'Young', 'Middle-aged', 'Senior' Learn how to use an int for bins parameter in the pd.cut() function to create the age groups.

2. Create a new feature 'HeartHealth' using the 'apply' function and the following conditions: Conditions: if thalach > 150 and chol < 240 Good, otherwise Poor

3. Create a new feature 'RiskFactor' using the following formula: RiskFactor = age * (chol + thalach)

```python


bins = [0, 35, 55, 100]  # Young: 0-35, Middle-aged: 36-55, Senior: 56-100
labels = ['Young', 'Middle-aged', 'Senior']

# Create 'AgeGroup' column using pd.cut()
df['AgeGroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

print(df)
```

```python
def heart_health(row):
    if row['thalach'] > 150 and row['chol'] < 240:
        return 'Good'
    else:
        return 'Poor'

df['HeartHealth'] = df.apply(heart_health, axis=1)

print(df)
```

```python
df['RiskFactor'] = df['age'] * (df['chol'] + df['thalach'])

print(df)
```

# Exercise 4: Drawing plots with Pyplot module functions


1. Draw a line plot of the 'trestbps' column for the first 50 rows. Line Plot of Trend of Resting Blood Pressure. Creat the data that is needed for the plot by slicing the DataFrame to get the first 50 rows of the 'trestbps' column


```python
import matplotlib.pyplot as plt
import pandas as pd
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
df = pd.read_csv('processed.cleveland.data', names=columns) 
data = df.head(50)

plt.figure(figsize=(8,5))
plt.plot(data.index, data['trestbps'], marker='o', linestyle='-', color='blue', label="Resting Blood Pressure")
plt.xlabel("index")
plt.ylabel("Resting Blood Pressure")
plt.title("Resting Blood Pressure for First 50 Patients")
plt.legend()
plt.show()
```

2. Draw a line plot of average 'trestbps' for each 'age' group. Line plot of average Trend of Resting Blood Pressure. It illustrates the average trend of Resting Blood Pressure (trestbps) across different age groups.

- X-Axis: Age

- Y-Axis: Average Resting Blood Pressure


NOTE: Use groupby to group the data by 'age' and calculate the mean of 'trestbps'.


The plot provides insight into how the average Resting Blood Pressure changes with increasing age.




```python
data = df.groupby('age')['trestbps'].mean()

plt.figure(figsize=(8,5))
plt.plot(data.index, data, marker='o', linestyle='-', color='blue', label="Resting Blood Pressure")
plt.xlabel("index")
plt.ylabel("Resting Blood Pressure")
plt.title("Resting Blood Pressure for each age")
plt.legend()
plt.show()
```

3. Draw a scatter plot showing the Relationship between Age and Risk Factor.

```python
import seaborn as sns

df['RiskFactor'] = df['age'] * (df['chol'] + df['thalach'])

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='RiskFactor', color='blue', alpha=0.6)

plt.xlabel("Age")
plt.ylabel("Risk Factor")
plt.title("Relationship between Age and Risk Factor")


plt.show()
```

4. Create a dataframe with the count of 'HeartHealth' for each 'AgeGroup' using the 'groupby' function. Learn about size() and unstack() functions.


```python
bins = [0, 35, 55, 100]  
labels = ['Young', 'Middle-aged', 'Senior']

df['AgeGroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

def heart_health(row):
    if row['thalach'] > 150 and row['chol'] < 240:
        return 'Good'
    else:
        return 'Poor'

df['HeartHealth'] = df.apply(heart_health, axis=1)

df = df.groupby(['AgeGroup', 'HeartHealth']).size().unstack()

print(df)
```

5. Draw a stacked bar plot showing the HeartHealth distribution within each Age group as the x-axis and the 'Good' and 'Poor' categories stacked for each age group. Add a legend to the plot.

```python
df.plot(kind='bar', stacked=True, figsize=(8, 5), color=['blue', 'orange'])

plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('HeartHealth Distribution within each Age Group')
plt.legend(title='HeartHealth')

plt.show()
```

6. Draw a histogram showing the distribution of patient ages. Set bins to 20 and color to 'skyblue'. Learn how to use the 'edgecolor' parameter to add a border to the bars.


- X-Axis: Age

- Y-Axis: Frequency (Number of individuals)




```python
df_age = pd.read_csv('processed.cleveland.data', names=columns)

plt.figure(figsize=(8, 5))
plt.hist(df_age['age'], bins=20, color='skyblue', edgecolor='black')  # Replace 'age' with the correct column name if different
plt.xlabel('Age')
plt.ylabel('Frequency (Number of individuals)')
plt.title('Distribution of Patient Ages')
plt.show()
```

7. Draw a pie chart showing the percentage distribution of the 'AgeGroup' column. Learn how to use the 'autopct' parameter to display the percentage value and the 'startangle' parameter to rotate the pie chart.



- Each slice of the pie corresponds to a specific age group (Young, Middle-aged, Senior).

- The 'skyblue', 'orange', and 'lightgreen' colors distinguish between the age groups.

- The startangle parameter (90 degrees) adjusts the starting position of the first slice.




```python
def categorize_age(age):
    if age < 35:
        return 'Young'
    elif 35 <= age < 55:
        return 'Middle-aged'
    else:
        return 'Senior'

df_age['AgeGroup'] = df_age['age'].apply(categorize_age)
age_group_counts = df_age['AgeGroup'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'lightgreen'])
plt.title('Percentage Distribution of Age Groups')
plt.show()
```
